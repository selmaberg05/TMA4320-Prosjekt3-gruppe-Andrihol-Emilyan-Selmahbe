```python
def secant(g, z0, z1, tol):
    max_iteratrions = 100

    for i in range(max_iteratrions):
        if np.abs(z0 - z1) < tol:
            return z1

        z = (z0 * g(z1) - z1 * g(z0)) / (g(z1) - g(z0))

        z0, z1 = z1, z

    raise ValueError(f"Konvergerer ikke etter {max_iteratrions} iterasjoner")
```

```python
import numpy as np

def g(z):
    return z + np.sin(z) + np.cos(z)
```

```python
rot = secant(g, 1, 2, 1e-4)
print(f"Rot = {rot:.4f}, g(rot) = {g(rot)}")
```
