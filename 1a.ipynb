For å løse initalverdiproblemet

$\frac{d^2y}{dx^2} = -4 · sin(2x), \quad y(0) = 0,  \quad y'(0)=2$

analytisk, er det mulig å integrere to ganger over problemet for å finne løsningen. Som kjent er

$y'' = -4 · sin(2x)$.

Det vil dermed være mulig å integrere over $y''$ for å løse for $y'$ med dens initialbetingelse.

$y' = \int y''(x)\,dx = \int -4 ·sin(2x)\,dx = -4 · (- \frac{1}{2}· cos(2x)) + C_1 = 2· cos(2x) + C_1$ 

Fra initalbetingelsene er som kjent, $y'(0)=2$, noe som gjør det mulig å løse ligningen for $C_1$.

$y'(0) = 2·cos(2·0) + C_1 = 2·cos(0) + C_1 = 2 + C_1 = 2$.

Dette medfører at $C_1 = 0$, altså er $y'=2· cos(2x)$. Denne prosessen gjentas for å løse for $y(x)$.

$y = \int\int y''(x)\,dx = \int y'\,dx = \int 2· cos(2x) \,dx = 2 · (\frac{1}{2} · sin(2x)) + C_2 = sin(2x) + C_2$

Igjen kan funksjonen løses for leddet $C_2$ med oppgitt initialbetingelse, $y(0) = 0$.

$y(0)  = sin(2·0) + C_2 = sin(0) + C_2 = 0 +C_2 = 0$

Igjen vil integrasjonsleddet falle bort og integrasjonen gir løsningen $y(x) = sin(2x)$.
