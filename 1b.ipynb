# Oppgave 1b)

Det er ønsket å skrive om den andreordens differensialligningen fra oppgave 1a) om til en førsteordens differensialligning for vektor 
$\vec{y}(x) = \begin{pmatrix}
y(x)\\
y'(x)
\end{pmatrix}.$

Det vil være hensiktmessig å sette $y(x) = y_1(x)$, og $y'(x) = y_2(x)$ for å skrive om til et førsteordens system. Dermed er 

$\vec{y}(x) = \begin{pmatrix} y(x)\\ y'(x) \end{pmatrix} = \begin{pmatrix} y_1(x)\\ y_2(x) \end{pmatrix}.$

Det gir dermed $y_1(x)' = y'(x) = y_2(x)$ og $y_2(x)' = y''(x) = -4 · sin(2x)$ som resulterer i føresteordens system

$\vec{y'}(x) = \begin{pmatrix} y_1'(x)\\ y_2'(x) \end{pmatrix} = \begin{pmatrix} y_2(x)\\ y''(x) \end{pmatrix} = \begin{pmatrix} y_2(x)\\ -4 · sin(2x) \end{pmatrix}.$

Dette kan kombineres med de kjente initalverdiene,

$\vec{y}(0) = \begin{pmatrix} y(0)\\ y'(0) \end{pmatrix} = \begin{pmatrix} 0\\ 2 \end{pmatrix}, $

som resulterer i

$\vec{y'}(x) = \begin{pmatrix} y_1'(x) = y_2(x)\\ y_2'(x) = - 4 · sin(2x) \\ y_1(0) = 0, \quad y_2(0) = 2 \end{pmatrix}$.
